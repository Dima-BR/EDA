# Exploratory Data Analysis (EDA) on "Car Price Prediction" Dataset

```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv("./data/CarPricePrediction/car_price_prediction_.csv")
df.head()
```

- **Missing values**: Check for any missing values in the dataset.

## Step 2: Descriptive Statistics


```python
# Descriptive statistics for numerical columns
df.describe()
```

```python
# توزيع الأسعار
plt.figure(figsize=(10, 6))
#sns.histplot(df['Price'], bins=30)
sns.histplot(df['Price'], bins=20, kde=True)
plt.title('Distribution of Car Prices')
plt.show()
```

```python
# العلاقة بين السنة والسعر
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Year', y='Price', data=df)
plt.title('Year vs. Price')
plt.show()
```

```python
# اختيار الأعمدة الرقمية فقط واستبعاد Car ID
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
numeric_columns = numeric_columns.drop('Car ID')  # استبعاد عمود Car ID

# مصفوفة الترابط
plt.figure(figsize=(10, 8))
sns.heatmap(df[numeric_columns].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix (Excluding Car ID)')
plt.show()
```

```python
# العلاقة بين الماركة والسعر
plt.figure(figsize=(12, 6))
sns.boxplot(x='Brand', y='Price', data=df)
plt.xticks(rotation=45)
plt.title('Brand vs. Price')
plt.show()
```
