# Exploratory Data Analysis (EDA) on "Student Performance Factors" Dataset

```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv("./data/StudentPerformanceFactors/StudentPerformanceFactors.csv")
df.head()
```

## Step 1: Understanding the Data
The first step is to get a basic understanding of the data:
- **Number of rows and columns**: Helps understand the dataset's size.
- **Quick glance at columns**: Identify the variables/features and their data types (numerical, text, etc.).
- **Missing values**: Check for any missing values in the dataset.

```python
# Check the number of rows and columns
print(df.shape)
```

## Overview of columns and their data types


```python
# نظرة سريعة على الأعمدة وأنواعها
df.info()
```

```python
df.columns
```

```python
# حساب القيم المفقودة في كل عمود
df.isnull().sum()
```

## Step 2: Descriptive Statistics


```python
# Descriptive statistics for numerical columns
df.describe()
```

```python
# رسم العلاقة بين ساعات الدراسة ونتيجة الامتحان
sns.scatterplot(x='Hours_Studied', y='Exam_Score', data=df)
plt.show()
```

```python
# رسم العلاقة بين الحضور ونتيجة الامتحان
sns.scatterplot(x='Attendance', y='Exam_Score', data=df)
plt.show()
```

```python
# توزيع الطلاب حسب الجنس
print(df['Gender'].value_counts())

# رسم بار جراف لتوزيع الجنس
df['Gender'].value_counts().plot(kind='bar')
plt.show()
```

```python
# رسم العلاقة بين نوع المدرسة ونتيجة الامتحان
sns.boxplot(x='School_Type', y='Exam_Score', data=df)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Sleep_Hours'], bins=30)
plt.title('Distribution of Sleep_Hours')
plt.xlabel('Sleep_Hours')
plt.ylabel('Frequency')
plt.show()
```

```python
# رسم العلاقة بين الدخل الأسري ونتيجة الامتحان
sns.scatterplot(x='Family_Income', y='Exam_Score', data=df)
plt.show()
```

```python
# رسم العلاقة بين مستوى تعليم الأهل ونتيجة الامتحان
sns.boxplot(x='Parental_Education_Level', y='Exam_Score', data=df)
plt.show()
```
